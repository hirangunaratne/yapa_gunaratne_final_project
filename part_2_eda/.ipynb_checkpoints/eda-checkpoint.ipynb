### 3.1.2 Analysing numerical variables

```python
# Get the correlation of all variables
forest_fire.corr()
```

```python
# Heatmap of correlation matrix
fig, ax = plt.subplots(figsize=(10,10))         
sns.heatmap(forest_fire.corr(),vmin=-1, vmax=1,cmap=sns.diverging_palette(h_neg=220, h_pos=10, n=21), linewidths=.5, ax=ax);
```

Our primary focus here is to understand how the feature variables correlate to the output variable 'area'.
Analysing the correlation matrix and seaborn heatmap, it is evident that all feature varables excluding 'RH' and 'rain' have a positive correlation whilst 'RH' and 'rain' have a negative correlation to the area of the forest fire. This can be considered a fare observation andboth 'RH' and 'rain' will limit the spread of forest fire

```python
# check for outliers via a box plot
ax = forest_fire.plot(kind='box',figsize=(20, 16))
```

Output variable **'area' seems to contain a couple of outliers** with values of over 600. These values may affect a generalised model. Hence it's sensible to exclude these outliers from the dataset

```python
# remove outliers in area variable
forest_fire_excl_outliers = forest_fire.loc[forest_fire.loc[:,'area']<600,:]
forest_fire_excl_outliers.shape
```

```python
# Determining the distribution of the data 
axes = forest_fire_excl_outliers.hist(figsize=(20, 15),bins=8)
```

```python
# view measures of central tendency of the dataset 
list_of_stats_list = [['Variable','Mean','Median','Mode']]

for v in ['X','Y','FFMC','DMC','DC','ISI','temp','RH','wind','rain','area']:
    list_of_stats_list.append([v,forest_fire_excl_outliers.loc[:,v].mean(),forest_fire_excl_outliers.loc[:,v].median(),forest_fire_excl_outliers.loc[:,v].mode()[0]])

headers = list_of_stats_list.pop(0)
central_tendency = pd.DataFrame(list_of_stats_list, columns=headers)    

central_tendency    
```

**'DC','RH','wind' and 'area'** seems to be **slightly more skewed** compared to the rest of the variavles . These can be transformed to build an effective predictive model, **skeweness of the other variavles are negligeable**

```python
# Scatter matrix of numerical columns
axes = sns.pairplot(forest_fire_excl_outliers)
```

Our primary focus here is to understand the linear relationship between feature variables and the output variable 'area'.
As per the pairplot, its evident that variables FFMC,DC, ISI, temp, RH, wind and rain has a notable linear relationship to the area of forest fire

### 3.3.3 Analysing non-numerical variables

```python
# freequency distribution by month
ax = forest_fire_excl_outliers.hist(column='area', by='month', sharex=True, sharey=True, figsize=(20, 12))
```

As per the data, **majarotiy of the forest fires** seems to have been observed **in** the months of **July, August** & **September**, with August & September being the most

```python
# freequency distribution by day
ax = forest_fire_excl_outliers.hist(column='area', by='day', sharex=True, sharey=True,figsize=(20, 12))
```

As per the data, there is **no clear indication of a specific set of days** where forest fires are more likely to occur 

**NOTE:**
Aim  of this project is to predict the area of the forest fire. Hence the yearly or weekly patterns of forest fire is irrelevant.
However, month and day of the forest fire may have a correlation to the area of th fire. 
Therefore dummy coding of these two variables will be used to understand the relationship

```python
# view caegory counts of month variable to decide on which category to drop when dummy coding
forest_fire_excl_outliers.loc[:,'month'].value_counts()
```

Most normal value for month **'aug' will be dropped**  when dummy coding

```python
# view caegory counts of day variable to decide on which category to drop when dummy coding
forest_fire_excl_outliers.loc[:,'day'].value_counts()
```

Most normal value for day **'sun' will be dropped**  when dummy coding

```python
# Dummy coding month & day
forest_fire_dummy_coded = pd.get_dummies(forest_fire_excl_outliers, columns=['month','day'])
forest_fire_dummy_coded.head()
```

```python
# Drop dummy coded columns for'aug'and 'sun'
forest_fire_dummy_coded = forest_fire_dummy_coded.drop(['month_aug','day_sun'], axis='columns')
forest_fire_dummy_coded.head()
```

```python
# Get the correlation of all variables
forest_fire_dummy_coded.corr()
```

```python
# Heatmap of correlation matrix
fig, ax = plt.subplots(figsize=(15,15))         
sns.heatmap(forest_fire_dummy_coded.corr(),vmin=-1, vmax=1,cmap=sns.diverging_palette(h_neg=220, h_pos=10, n=21), linewidths=.5, ax=ax);
```

Our primary focus here is to understand how the feature variables correlate to the output variable 'area'.
Analysing the correlation matrix and seaborn heatmap for the dummy coded variables, it is evident that 'month_sept' and 'day_sat' have a notable positive correlation whilst 'month_mar', 'day_fri, and 'day_thu' have a notable negative correlation to the area of the forest fire. Other dummy coded variables show a less notable correlation
